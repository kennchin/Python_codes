```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/kennchin/python_missing_data_example/master/train.csv"
titanic = pd.read_csv(url,index_col='PassengerId')
```

```python
titanic.head()
```

```python
#find out the number of missing value for each predictor
print (titanic.count())
titanic.isnull().sum()
```

Note:
    - There is a total of 891 rows in our dataset
    - The predictor Age has 177 missing values, Cabin has 687 missing  values, and Embarked has 2 missing values.

```python
#Delete the row if there is a missing value in any of the predictors
titanic_drop = titanic.dropna()
titanic_drop
```

```python
#Delete row with missing value in the predictor Age.
titanic_drop_age = titanic[titanic.Age.notnull()]
```

```python
titanic_drop_age
```

Note:
    - Row 18 has been deleted since there is a missing value. (Our       intention)
    - Row 19 is intact since there is a missing value in the predictor Cabin.

```python
#Impute missing values with either mean or median or mode
#Let's look at the distribution of Age
titanic.hist(column="Age")
```

Note:
    - The data for predictor Age seems to be skewed to the right meaning that median is appropriate as a way of summarizing the data since it's robust to skewness.

```python
#Let's impute our missing values of Age with the median.
print (titanic.Age.median())
titanic.fillna(titanic.Age.median(), inplace=True)
```
