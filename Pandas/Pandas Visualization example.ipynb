```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# using csv parser
countries = []

with open('Food.csv',encoding='latin1') as f:
    reader = csv.reader(f)
    for row in reader:
        countries.append([row[1],row[3],row[5],row[7]])
        
```

```python
#using pandas to read csv file
A = pd.read_csv("Food.csv",encoding="latin-1")
```

```python
#Subsetting list to only 4 column of interests: Country, Item, Element, and 1961
Asub = A[['Country','Item','Element','Y1961']]
```

```python
Asub.head()
```

```python
# Subsetting Element column with Food supply quantity as variable of interests
B = Asub[Asub.Element=='Food supply quantity (kg/capita/yr)']
```

```python
# Subsetting list with Item column focusing on Wheat and products
C = B[B.Item=='Wheat and products']
C_sorted = C.sort_values(by='Country')
```

```python
C.plot(kind="bar",x='Country',y="Y1961")
```

```python
C.boxplot(column="Y1961", by="Country")
```

```python
#The above bar chart is too messy since there are so many countries; let's just focus on the first 10 countries 
C_small = C_sorted[:10]
```
